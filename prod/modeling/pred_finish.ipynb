```python
for year in range(2010,2023): 
    n_rounds = f1.get_event_schedule(year=year, include_testing=False).RoundNumber.max()
    for round in range(1, n_rounds+1):
        for session in ["Q"]:
            session.load()
            laps = session.laps

            fastest = pd.DataFrame()
            for num in timing.get_driver_numbers(session):
                fastest = pd.concat([fastest, laps[laps["DriverNumber"] == num].pick_fastest()], axis=1)
            fastest = fastest.T
            cols = ["Driver", "DriverNumber", "LapTime", "Sector1Time", "Sector2Time", "Sector3Time", "SpeedFL", "SpeedI1", "SpeedI2", "SpeedST"]
            fastest = fastest[cols]
            fastest.loc[:, fastest.columns.str.contains("Time")] = fastest.select_dtypes("timedelta64") / np.timedelta64(1, "s")
            fastest.columns[2:] = fastest.columns[2:] + "_" + session
```
